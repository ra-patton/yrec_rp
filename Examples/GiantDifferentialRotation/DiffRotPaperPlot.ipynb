```python
from astropy.io import ascii
from astropy.table import Table
import matplotlib.pyplot as plt
import numpy as np
```

```python
#making Paper plot
rot3I0cz0=Table.read('Test_3Msolar_GS_rot_gray_OPALSCV_SF2_GS98_OP_CF10_RGp2diffrotI0.track',
                 format='ascii') #, delimiter=' ', data_start=1, header_start=0)
rot3lastI0cz0=Table.read('Test_3Msolar_GS_rot_gray_OPALSCV_SF2_GS98_OP_CF10_RGp2diffrotI0.last',
                 format='ascii', delimiter=' ', data_start=7, header_start=6)

rot3I1cz0=Table.read('Test_3Msolar_GS_rot_gray_OPALSCV_SF2_GS98_OP_CF10_RGp2diffrotI1.track',
                 format='ascii') #, delimiter=' ', data_start=1, header_start=0)
rot3lastI1cz0=Table.read('Test_3Msolar_GS_rot_gray_OPALSCV_SF2_GS98_OP_CF10_RGp2diffrotI1.last',
                 format='ascii', delimiter=' ', data_start=7, header_start=6)

rot3I3cz2=Table.read('Test_3Msolar_GS_rot_gray_OPALSCV_SF2_GS98_OP_CF10_RGp2diffrotI3cz2.track',
                 format='ascii') #, delimiter=' ', data_start=1, header_start=0)
rot3lastI3cz2=Table.read('Test_3Msolar_GS_rot_gray_OPALSCV_SF2_GS98_OP_CF10_RGp2diffrotI3cz2.last',
                 format='ascii', delimiter=' ', data_start=7, header_start=6)
```

```python
print(rot3I3cz0['Y_cen'][-1], rot3I0cz2['Y_cen'][-1])
```

```python
convrot3I0cz0=np.where(rot3lastI0cz0['C']=='T')
convrot3I1cz0=np.where(rot3lastI1cz0['C']=='T')
convrot3I3cz2=np.where(rot3lastI3cz2['C']=='T')
```

```python
release_rcParams = {
    'font.family': 'serif',
    'font.serif': 'Liberation Serif',
    'axes.labelsize': 9,
    'legend.fontsize': 9,
    'font.size': 9,
    'xtick.labelsize': 9,
    'ytick.labelsize': 9,
    'figure.dpi': 300,
    'figure.figsize': [3.4,2.5], # change the second number for aspect, but the first is set to just shy of AAS column widt
    'text.usetex': True,
    'xtick.minor.visible': True,
    'ytick.minor.visible': True,
    'figure.autolayout': True,
    
}

```

```python
#activate by running:
plt.rcParams.update(release_rcParams)
```

```python

plt.plot(10**rot3lastI1cz0['MASS'][convrot3I1cz0][0:1]/1.99e33, 2*3.14159/(60*60*24*rot3lastI1cz0['OMEGA'][convrot3I1cz0][0:1]),c='pink', label='Convective')

plt.plot(10**rot3lastI1cz0['MASS']/1.99e33, 2*3.14159/(60*60*24*rot3lastI1cz0['OMEGA']), label='Solid Body', zorder=1)
plt.plot(10**rot3lastI0cz0['MASS']/1.99e33, 2*3.14159/(60*60*24*rot3lastI0cz0['OMEGA']),  label='Solid Body CZ', zorder=1)
plt.plot(10**rot3lastI3cz2['MASS']/1.99e33, 2*3.14159/(60*60*24*rot3lastI3cz2['OMEGA']),  label='Differential SCZ', zorder=1)



plt.scatter(10**rot3lastI1cz0['MASS'][convrot3I1cz0][0:1]/1.99e33, 2*3.14159/(60*60*24*rot3lastI1cz0['OMEGA'][convrot3I1cz0][0:1]),s=1, c='pink', zorder=2)

plt.scatter(10**rot3lastI1cz0['MASS'][convrot3I1cz0]/1.99e33, 2*3.14159/(60*60*24*rot3lastI1cz0['OMEGA'][convrot3I1cz0]), s=1,  c='pink')

plt.scatter(10**rot3lastI0cz0['MASS'][convrot3I0cz0]/1.99e33, 2*3.14159/(60*60*24*rot3lastI0cz0['OMEGA'][convrot3I0cz0]), s=1, c='pink')
plt.scatter(10**rot3lastI3cz2['MASS'][convrot3I3cz2]/1.99e33, 2*3.14159/(60*60*24*rot3lastI3cz2['OMEGA'][convrot3I3cz2]), s=1, c='pink')

plt.xlabel(r'Mass Coordinate [M$_\odot$]')
plt.ylabel('P$_{rot}$ [days]')

plt.legend(frameon=False)


plt.xlim(0, 3)
plt.ylim(1000, 1)
plt.yscale('log')
plt.savefig('3MsunRGp2wczimpI123_decmassJenstyle.jpg')
```
